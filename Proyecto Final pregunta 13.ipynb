# Probabilidades del Trade Price del Bond Id 6968

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

b = pd.read_csv('/Users/juliocastillo/Desktop/IE/Tercer Semestre/Data Science/Proyecto Final/bonds.csv')
bybond = b.groupby('bond_id')
bybond['bond_id'].count()
maximo = max(bybond['bond_id'].count())
maximo

bybond.bond_id.count()[bybond.bond_id.count() == 12129]

b.trade_size[b.bond_id == 6968].describe()
b.trade_price[b.bond_id == 6968].describe()

# Primer cuartil = [0 ;  10,000]
# Cuarto cuartil = [30,000 ; 5,000,001]

# Cantidad de trade prices que estan en el bond id = 6968 y en los cuartiles del trade size.

s1 = (b.trade_price[(b.bond_id == 6968)&(b.trade_size <= 10000)]).count()
s2 = (b.trade_price[(b.bond_id == 6968)&(b.trade_size >= 10000)&(b.trade_size <= 20000)]).count()
s3 = (b.trade_price[(b.bond_id == 6968)&(b.trade_size <= 30000)&(b.trade_size >= 20000)]).count()
s4 = (b.trade_price[(b.bond_id == 6968)&(30000 <= b.trade_size)]).count()

#Probabilidad que el trade_price en el primer cuartil sea igual al promedio:
#Promedio del precio en el primer cuartil. 
s_q1 = b.trade_price[(b.bond_id == 6968)&(b.trade_size <= 10000)].sum()/s1
s_q1

r1 = s_q1 - 0.5
r2 = s_q1 + 0.5

c_s_q1 = b.trade_price[(b.bond_id == 6968)&(b.trade_size <= 10000)]
c1 = c_s_q1[(r1 <= c_s_q1)&(c_s_q1 <= r2)].count()
p1 = float(c1)/float(s1)
p1

#Probabilidad que el trade_price en el segundo cuartil sea igual al promedio:
s_q2 = b.trade_price[(b.bond_id == 6968)&(b.trade_size >= 10000)&(b.trade_size <= 20000)].sum()/s2
s_q2

r3 = s_q2 - 0.5
r4 = s_q2 + 0.5

c_s_q2 = b.trade_price[(b.bond_id == 6968)&(b.trade_size <= 10000)&(b.trade_size <= 20000)]
c2 = c_s_q1[(r3 <= c_s_q2)&(c_s_q2 <= r4)].count()
p2 = float(c2)/float(s2)
p2

#Probabilidad que el trade_price en el tercer cuartil sea igual al promedio:
s_q3 = b.trade_price[(b.bond_id == 6968)&(b.trade_size <= 30000)&(b.trade_size >= 20000)].sum()/s3
s_q3

r5 = s_q3 - 0.5
r6 = s_q3 + 0.5

c_s_q3 = b.trade_price[(b.bond_id == 6968)&(b.trade_size <= 30000)&(b.trade_size >= 20000)]
c3 = c_s_q3[(r5 <= c_s_q3)&(c_s_q3 <= r6)].count()
p3 = float(c3)/float(s3)
p3

#Probabilidad que el trade_price en el cuarto cuartil sea igual al promedio:
s_q4 = b.trade_price[(b.bond_id == 6968)&(30000 <= b.trade_size)].sum()/s4
s_q4

r7 = s_q4 - 0.5
r8 = s_q4 + 0.5

c_s_q4 = b.trade_price[(b.bond_id == 6968)&(30000 <= b.trade_size)]
c4 = c_s_q4[(r7 <= c_s_q4)&(c_s_q4 <= r8)].count()
p4 = float(c4)/float(s4)
p4

#Crosstab

raw_data = {'Cantidad de transacciones': [s1,s2,s3,s4], 'Precio promedio': [s_q1,s_q2,s_q3,s_q4], 'Probabilidades': [p1,p2,p3,p4]}
df = pd.DataFrame(raw_data, index = ['Cuartil 1', 'Cuartil 2', 'Cuartil 3','Cuartil 4'], columns = ['Cantidad de transacciones','Precio promedio','Probabilidades'])
df
```
