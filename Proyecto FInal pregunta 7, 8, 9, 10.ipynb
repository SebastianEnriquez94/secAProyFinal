## ¿En promedio, qué tiempo de madurez tienen los bonos más caros y los bonos mas baratos?, ¿Qué porcentaje de los bonos baratos tienen un tiempo de madurez mayor a t años?

```python
bonds.trade_price.describe()
#bonos baratos $99.433 o menes (primer cuartil)
#bonos caros $107.11 o más (último cuartil)
low_price = bonds.trade_price < 99.50
high_price = bonds.trade_price > 107.00
```

```python
bonds.time_to_maturity[low_price].mean()
#los bonos mas baratos tienen un tiempo de madurez promedio de 7.9655 años
```

```python
bonds.time_to_maturity[high_price].mean()
#los bonos mas caros tienen un tiempo de madurez promedio de 8.6504 años
```

```python
allprices = bonds.trade_price.count() #conteo de todos los datos
```

```python
total = bonds.time_to_maturity.count()
def maturity_time(time):
    less_than_t = bonds.time_to_maturity <= time
    price_count = low_price[less_than_t].count()
    porcentaje = (price_count/total)*100
    return porcentaje

m_time = input("que tiempo de madurez desea? ")
maturity_time(m_time) 
```

## ¿Qué porcentaje de los bonos con tasas de interes altas (current_coupon) serán liquidados en, máximo, t años? 

```python
bonds.time_to_maturity.describe()
#intereses altos 6.5 o mas
```

```python
total_high = bonds.time_to_maturity[bonds.current_coupon >= 6.5]
total_high.count()
#193337 bonos poseen interes mayor que 6.5%

def mttime(time):
    high_this_year = bonds.time_to_maturity[(bonds.current_coupon >= 6.5) & (bonds.time_to_maturity <= time)]
    porc = (high_this_year.count()/total_high.count())*100
    porcb = porc
    return porcb

m_time = input("que tiempo de madurez desea? ")
mttime(m_time)
```

## ¿Qué porcentaje de los volumenes de las últimas dos transacciones han disminuido si se presenta un tiempo de retraso mayor a t segundos?

```python
#ultima transaccion
bonds.trade_size.describe()
#transacciones de gran volumen: 100,000 o más 
big_size = bonds.trade_size >100000
big_size. describe() #762678 transacciones con volumenes mayores a 100,000

#penultima transacción
bonds.trade_size_last1.describe()
#transacciones de gran volumen: 100,000 o más 
old_big_size = bonds.trade_size >100000
old_big_size.describe() #762678 transacciones con volumenes mayores a 100,000
```

```python
#ultima transacción 
def delay(dtime):
    big_delay_transactions = (bonds.reporting_delay[(bonds.trade_size >= 100000) & (bonds.reporting_delay <= dtime)]).count()
    return big_delay_transactions

#penúltima transacción   
def old_delay(odtime):
    old_big_delay_transactions = (bonds.reporting_delay[(bonds.trade_size_last1 >= 100000) & (bonds.reporting_delay <= odtime)]).count()
    return old_big_delay_transactions

delayt = input("que tiempo de retraso desea saber? ")
delay(delayt)
old_delay(delayt)

difference = old_delay(delayt)-delay(delayt)
difference/(bonds.reporting_delay.count())*100
```
