## ¿Cual es la probabilidad que un bono de alto o bajo interes pueda ser liquidado en cualquier momento?

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

b = pd.read_csv('bonds prices.csv')
```

```python
sns.boxplot(b.current_coupon)
sns.boxplot(b.trade_price)
```

```python
b.current_coupon.describe()
```

```python
b.trade_price.describe()
```

```python
pd.crosstab((b.current_coupon > 6.5), (b.is_callable == 1))
```

```python
def coupon_callable(couponx):
    are_callable = b.is_callable[(b.is_callable == 1)]
    mayores = are_callable[(b.current_coupon >= 6.5)]
    menores = are_callable[(b.current_coupon < 6.5)]
    if couponx >= 6.5:
        x = float(mayores.count())/float(are_callable.count())
    else:
        x = float(menores.count())/float(are_callable.count())
    return x


coupon_callable(8) #0.6401854859590895
coupon_callable(3) #0.35981451404091047
```

```python
b.groupby(['current_coupon','trade_price']).count()
```

## Probabilidad de un bono bajo es 0.35981451404091047
