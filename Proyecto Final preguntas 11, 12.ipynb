```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from __future__ import division

bonds = pd.read_csv('bonds prices.csv')
```

```python
#cual es la probabilidad de que al liquidar el bono se haya obtenido una ganancia del 50%?
#asumiendo que el interes (current_coupon) es anual
interes = (bonds.current_coupon)/100
tiempo = bonds.time_to_maturity
price = bonds.trade_price
ganancia = (interes * price) * tiempo

porcentaje_ganancia = ganancia/price
cincuentaomas = sum(porcentaje_ganancia >= 0.5)
(cincuentaomas)
prob = (cincuentaomas)/(price.count())
prob
```

```python
#Cual es la probabilidad que un bono de X tipo de transacciÃ³n pueda ser liquidado en cualquier momento (is_callable)
total_call = bonds.is_callable[(bonds.is_callable == 1)]
def type_callable(typet):
    by_type = bonds.trade_type[(bonds.trade_type == typet)]
    callable_bonds = by_type[(bonds.is_callable == 1)]
    prob_call = (callable_bonds.count())/total_call.count()
    return prob_call

type_callable(2) #0.21151990207847846
type_callable(3) #0.38762563849069037
type_callable(4) #0.40085445943083114
```
